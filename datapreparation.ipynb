```python
routeData = pd.DataFrame(columns=['trainNumber','departureTime', 'stationCode', 'haltTime', 'dayCount', 'routeNumber',
       'distance', 'arrivalTime', 'stationName', 'stnSerialNumber'])
```

```python
for train in trains["trainNumber"]:
    try:
        trainData = trains[trains["trainNumber"] == train]["stationList"]
        # print(train_data)
        for route in trainData:
            df = pd.DataFrame(route)
            df.insert(0, 'trainNumber', train)
        routeData = pd.concat([routeData, df], ignore_index=True)

    except: 
        continue    

```

```python
routeData[routeData["trainNumber"]== 22655]
```

```python
stationData = routeData[['stationCode', 'stationName']]
```

```python
stationData.drop_duplicates(subset={'stationCode', 'stationName'}, inplace=True)
```

```python
stationData.reset_index(inplace=True,drop=True)
```

```python
routeData.drop(columns=['routeNumber'])
```

```python
import numpy as np
routeData.replace("--", np.NaN, inplace=True)
```

```python
routeData['stationTo'] = routeData.groupby('trainNumber')['stationCode'].shift(-1)
```

```python
routeData['distance'] = routeData.groupby('trainNumber')['dayCount'].shift(-1).astype(np.int32) - routeData['routeNumber'].astype(np.int32)
```
