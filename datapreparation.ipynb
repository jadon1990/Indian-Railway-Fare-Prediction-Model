#### Prepare Route Data

```python
trains["stationList"][0][0], trains["stationList"][0][-1]
```

```python
routeData = pd.DataFrame(columns=['trainNumber','departureTime', 'stationCode', 'haltTime', 'dayCount', 'routeNumber',
       'distance', 'arrivalTime', 'stationName', 'stnSerialNumber'])

for train in trains["trainNumber"]:
    try:
        trainData = trains[trains["trainNumber"] == train]["stationList"]
        # print(train_data)
        for route in trainData:
            df = pd.DataFrame(route)
            df.insert(0, 'trainNumber', train)
        routeData = pd.concat([routeData, df], ignore_index=True)

    except: 
        continue    

```

```python
stationData = routeData[["stationCode", "stationName"]]
```

```python
stationData.drop_duplicates(subset=['stationCode', 'stationName'], inplace=True)
```

```python
stationData[stationData["stationCode"] == "SONI"]
```

```python
routeData[routeData["trainNumber"]== 22655]
```

```python
routeData.drop(columns=['routeNumber'])
```

```python
import numpy as np
routeData.replace("--", np.NaN, inplace=True)
```

```python
routeData['stationTo'] = routeData.groupby('trainNumber')['stationCode'].shift(-1)
```

```python
routeData['distance'].replace(np.NaN, 0 , inplace=True)
```

```python
routeData['distance'] = routeData['distance'].astype(np.int32)
routeData['distance'] = routeData.groupby('trainNumber')['distance'].shift(-1) - routeData['distance']
```

```python
routeData = routeData[['trainNumber', 'stnSerialNumber' , 'stationCode', 'stationTo', 'distance', 'departureTime', 'arrivalTime', 'haltTime', 'dayCount']]
```

```python
routeData = routeData.merge(trains[['trainNumber' ,'timeStamp']], on='trainNumber')
```

```python
routeData['departureTime'] = routeData['timeStamp'].astype('str').str.split(" ")[0][0] + " " + routeData['departureTime'].astype('str')
routeData['arrivalTime'] = routeData['timeStamp'].astype('str').str.split(" ")[0][0] + " " + routeData['arrivalTime'].astype('str')
```

```python
import re
routeData['arrivalTime'] = routeData['arrivalTime'].apply(lambda x: re.sub('nan$','00:00',x))
routeData['departureTime'] = routeData['departureTime'].apply(lambda x: re.sub('nan$','00:00',x))
```

```python
routeData['arrivalTime1'] = pd.to_datetime(routeData['arrivalTime'], format='%Y-%m-%d %H:%M')
routeData['departureTime1'] = pd.to_datetime(routeData['departureTime'], format='%Y-%m-%d %H:%M')
```

```python
routeData['dayCount'] = routeData['dayCount'].astype('int')
```

```python
routeData['dayCount'] = routeData['dayCount'] - 1 
```

```python
routeData['dayCount'] = routeData.groupby('trainNumber')['dayCount'].shift(-1)
```

```python
routeData['journeyTime'] = (pd.to_datetime(routeData.groupby('trainNumber')['arrivalTime'].shift(-1), format='%Y-%m-%d %H:%M') - pd.to_datetime(routeData['departureTime'], format='%Y-%m-%d %H:%M')) / pd.Timedelta(minutes=1)
```

```python
routeData = routeData[routeData['journeyTime'] > 0]
```

```python
routeData.drop(columns=['arrivalTime1', 'departureTime1'], inplace=True)
```

```python
routeData['averageSpeed'] = routeData['distance']/(routeData['journeyTime']/60)
```

```python
routeData['trainStatus'] = np.where(routeData['averageSpeed'] > 50, 1, 0)
```

```python

def calculateFare(distanc, time, isSuperFast):
    distanc = float(distanc)
    time = float(time) 
  
    
    baseFare = {
        'base' : 50,
        'superFase' : 100,
        'perKM' : 50,
        'time' : 25
    }

    if isSuperFast > 0:
        isSuperFastChage = baseFare['superFase']
    else: 
        isSuperFastChage = 0


    fare = baseFare['base'] + (baseFare['time']*time) + (baseFare['perKM']*distanc) + isSuperFastChage
    return fare  
```

```python
calculateFare(1, 1, 0)
```

```python
baseFare = {
        'base' : 50,
        'superFase' : 100,
        'perKM' : 50,
        'time' : 25
    }
```

```python
routeData['fare'] = routeData['distance']*baseFare['perKM'] + routeData['journeyTime']*baseFare['time']
routeData['fare'] = routeData['fare'] + np.where(routeData['trainStatus'] == 1, baseFare['superFase'], 0)
# (baseFare['superFase'] if routeData['trainStatus'] else 0)
```

```python
routeData
```

```python
final_df = routeData[['distance', 'journeyTime', 'trainStatus', 'fare']]
```

```python
final_df.info()
```

```python
x = final_df.drop(columns=['fare'], axis=1)
y = final_df['fare']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2)
```

```python
## Import ML Libraries
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error
from xgboost import XGBRegressor
```

```python
pipe = Pipeline(steps=[
    ('step1', StandardScaler()),
    ('step', XGBRegressor(n_estimators = 50, max_depth=25, random_state=1))
])
```

```python
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
